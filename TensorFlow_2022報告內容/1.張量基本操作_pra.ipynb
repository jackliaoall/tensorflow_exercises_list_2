TF變數類型主要有三種，tf.constant、tf.Variable以及tf.placeholder。

1.tf.constant: 一個constant的類別，你可以放入各式的資料型態，做一些基礎運算，Ex: tf.add 等等。

2.tf.Variable: 放置可學習的變數或者說將可求導的變數，例如: Neural Network的wieght或者bias。

3.tf.placeholder: 在TF2.0中，已經不使用這個變數。

**三，維度變換**

維度變換相關函數主要有tf.reshape, tf.squeeze, tf.expand_dims, tf.transpose.

1.tf.reshape: 可以改變張量的形狀。

2.tf.squeeze: 可以減少維度。

3.tf.expand_dims: 可以增加維度。

4.tf.transpose: 可以交換維度。

```python
a = tf.random.uniform(shape=[1,3,3,2], minval=0,maxval=255,dtype=tf.int32)
tf.print(a.shape)
tf.print(a)
```

```python
#改成[3,6]形狀的張量
a = tf.reshape(a, [3, 6])
a
```

```python
#改回成[1,3,3,2]形狀的張量
a = tf.reshape(a, [1,3,3,2])
a
```

```python
#如果張量在某個維度上只有一個元素，利用tf.squeeze可以消除這個維度。
a = tf.squeeze(a)
a
```

```python
#在第0維插入長度為1的一個維度
a = tf.expand_dims(a, axis=0)
a
```

```python
# Batch,Height,Width,Channel => [4,60,60,100]
a = tf.random.uniform(shape=[100,600,600,4],minval=0,maxval=255,dtype=tf.int32)
tf.print(a.shape)

#用tf.transpose轉換成Channel,Height,Width,Batch
s = tf.transpose(a, [3,2,1,0])
tf.print(s.shape)
```
